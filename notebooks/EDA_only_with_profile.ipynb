```python
from xgboost import XGBClassifier
```

```python
#create model
XGBoost_model = XGBClassifier(random_state = 42)
```

```python
#train the model
XGBoost = XGBoost_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import (
    confusion_matrix, classification_report,
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, RocCurveDisplay
)
import matplotlib.pyplot as plt

def eval_metric(model, X_train, y_train, X_test, y_test):
    # Predict labels and probabilities
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    y_test_proba = model.predict_proba(X_test)[:, 1]

    # --- Test Set Evaluation ---
    print("🔍 Test Set Evaluation")
    print("------------------------")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_test_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_test_pred))
    print("Metrics:")
    print(f"Accuracy : {accuracy_score(y_test, y_test_pred):.3f}")
    print(f"Precision: {precision_score(y_test, y_test_pred):.3f}")
    print(f"Recall   : {recall_score(y_test, y_test_pred):.3f}")
    print(f"F1-score : {f1_score(y_test, y_test_pred):.3f}")
    print(f"ROC AUC  : {roc_auc_score(y_test, y_test_proba):.3f}")
    print()

    # --- Train Set Evaluation ---
    print("🧪 Train Set Evaluation")
    print("-------------------------")
    print("Confusion Matrix:")
    print(confusion_matrix(y_train, y_train_pred))
    print("\nClassification Report:")
    print(classification_report(y_train, y_train_pred))
    print("Metrics:")
    print(f"Accuracy : {accuracy_score(y_train, y_train_pred):.3f}")
    print(f"Precision: {precision_score(y_train, y_train_pred):.3f}")
    print(f"Recall   : {recall_score(y_train, y_train_pred):.3f}")
    print(f"F1-score : {f1_score(y_train, y_train_pred):.3f}")
    # No AUC for train set unless you also compute train probabilities

    # --- ROC Curve ---
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.title("ROC Curve - Test Set")
    plt.show()
```

```python
eval_metric(XGBoost_model, X_train, y_train, X_test, y_test)
```
